```python

def get_addrs(x):
    x1 = x.split(' ')
    return ' '.join(x1[0:5])
JungGuaddress['전처리 주소'] = JungGuaddress['주소'].apply(get_addrs)
JungGuaddress.head(200)
```

```python
geo_local = Nominatim(user_agent='South Korea')

def geocoding_lat(JungGuaddress):
    try:
        geo = geo_local.geocode(JungGuaddress)
        return geo.latitude
    except:
        return 0

def geocoding_lon(JungGuaddress):
    try:
        geo = geo_local.geocode(JungGuaddress)
        return geo.longitude
    except:
        return 0
JungGuaddress['위도'] = JungGuaddress['전처리 주소'].apply(geocoding_lat)
JungGuaddress['경도'] = JungGuaddress['전처리 주소'].apply(geocoding_lon)
JungGuaddress.head(200)
```

```python
nulls = JungGuaddress[JungGuaddress['위도'] == 0]
null_index = list(nulls.index)
null_index
```

```python
for i in null_index:
    JungGuaddress['위도'][i] = geocoding_lat(JungGuaddress['전처리 주소'][i])
    JungGuaddress['경도'][i] = geocoding_lon(JungGuaddress['전처리 주소'][i])
JungGuaddress[JungGuaddress['위도'] ==0]
```
