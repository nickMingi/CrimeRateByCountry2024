```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Haversine 공식을 사용하여 두 지점 사이의 거리 계산
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # 지구의 반지름 (단위: km)
    
    # 라디안 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도와 경도의 차이
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine 공식 계산
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # 거리 반환
    return R * c

# 데이터 프레임 생성 (관서명, 위도, 경도)
data = {
    '관서명': ['창동', '신창', '방학', '신방학', '도봉1', '도봉2', '쌍문', '숭미'],
    'Latitude': [37.652288, 37.643522, 37.664465, 37.658501, 37.679666, 37.683987, 37.657568, 37.647487],
    'Longitude': [127.051722, 127.034575, 127.039996, 127.027144, 127.043351, 127.049283, 127.038029, 127.025580]
}
df = pd.DataFrame(data)

# 모든 경우의 수에 대해 최소 거리와 관서 간의 이동 경로를 저장
min_distances = {}
min_distance_paths = {}

for i, row1 in df.iterrows():
    for j, row2 in df.iterrows():
        if i != j:
            distance = haversine(row1['Latitude'], row1['Longitude'],
                                 row2['Latitude'], row2['Longitude'])
            if row1['관서명'] not in min_distances or distance < min_distances[row1['관서명']]:
                min_distances[row1['관서명']] = distance
                min_distance_paths[row1['관서명']] = f"{row1['관서명']} -> {row2['관서명']}"
            elif distance == min_distances[row1['관서명']]:
                min_distance_paths[row1['관서명']] += f", {row1['관서명']} -> {row2['관서명']}"

# 중복되지 않는 최소 거리 값을 출력
unique_min_distances = sorted(set(min_distances.values()))

# 각 관서별 최소 거리 및 이동 경로 출력
for station in df['관서명']:
    print(f"*** {station}에서 다른 관서로의 최소 거리 및 이동 경로 ***")
    for distance in unique_min_distances:
        if min_distances[station] == distance:
            print(f"최소 거리: {distance:.2f} km")
            print(f"이동 경로: {min_distance_paths[station]}")
            print()
            break
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Haversine 공식을 사용하여 두 지점 사이의 거리 계산
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # 지구의 반지름 (단위: km)
    
    # 라디안 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도와 경도의 차이
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine 공식 계산
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # 거리 반환
    return R * c

# 데이터프레임 생성
data = {
    '관서명': ['당현', '월계', '화랑', '노원역', '불암'],
    'Latitude': [37.642026, 37.633666, 37.622542, 37.655007, 37.652429],
    'Longitude': [127.068365, 127.059276, 127.074550, 127.066943, 127.077329]
}
df = pd.DataFrame(data)

# 각 관서 간의 최단 거리를 저장할 딕셔너리
min_distances = {}

# 각 관서마다 다른 관서로의 거리를 계산하여 최단 거리와 두 번째로 최소인 거리를 저장
for i in range(len(df)):
    distances = []  # 각 관서 간의 거리를 저장할 리스트
    for j in range(len(df)):
        if i != j:  # 자기 자신과의 거리는 계산하지 않음
            distance = haversine(df.iloc[i]['Latitude'], df.iloc[i]['Longitude'],
                                 df.iloc[j]['Latitude'], df.iloc[j]['Longitude'])
            distances.append((df.iloc[j]['관서명'], distance))  # 관서명과 거리를 튜플로 저장
    distances.sort(key=lambda x: x[1])  # 거리를 기준으로 오름차순 정렬
    min_distances[df.iloc[i]['관서명']] = {'최단 거리1': distances[0], '최단 거리2': distances[1]}

# 결과 출력
for k, v in min_distances.items():
    print(f"*** {k} 관서의 최단 거리 ***")
    print(f"최단 거리1: {v['최단 거리1'][1]:.2f} km ({v['최단 거리1'][0]})")
    print(f"최단 거리2: {v['최단 거리2'][1]:.2f} km ({v['최단 거리2'][0]})\n")
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Haversine 공식을 사용하여 두 지점 사이의 거리 계산
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # 지구의 반지름 (단위: km)
    
    # 라디안 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도와 경도의 차이
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine 공식 계산
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # 거리 반환
    return R * c

# 데이터프레임 생성
data = {
    '관서명': ['당현', '월계', '화랑', '노원역', '불암'],
    'Latitude': [37.642026, 37.633666, 37.622542, 37.655007, 37.652429],
    'Longitude': [127.068365, 127.059276, 127.074550, 127.066943, 127.077329]
}
df = pd.DataFrame(data)

# 모든 관서 간의 거리를 저장할 딕셔너리
distances = {}

# 모든 관서 간의 거리 계산
for i in range(len(df)):
    distances[df.iloc[i]['관서명']] = {}  # 각 관서별로 거리를 저장할 딕셔너리 초기화
    for j in range(len(df)):
        if i != j:  # 자기 자신과의 거리는 계산하지 않음
            distance = haversine(df.iloc[i]['Latitude'], df.iloc[i]['Longitude'],
                                 df.iloc[j]['Latitude'], df.iloc[j]['Longitude'])
            distances[df.iloc[i]['관서명']][df.iloc[j]['관서명']] = distance  # 관서명과 거리를 저장

# 결과 출력
for k, v in distances.items():
    print(f"*** {k} 관서의 거리 ***")
    for target, distance in v.items():
        print(f"{target}: {distance:.2f} km")
    print()
```
